```python
import pandas as pd
import csv
import datetime as dt
```

```python
csv_file_1="Resources/mass_2016 (1).csv"
mass_1971_1981 = pd.read_csv(csv_file_1)
mass_1971_1981.head()
```

```python
drop_mass=mass_1971_1981[["Title","Location","Number of Victim Fatalities","Number of Victims Injured","Total Number of Victims","Date","Average Shooter Age","Shooter Sex","Shooter Race"]].copy()
drop_mass.head()
```

```python
mass_1971_1981_rename=drop_mass.rename(columns={"Title":"Shooting_Case","Location":"Shooting_Location","Number of Victim Fatalities":"Fatalities","Number of Victims Injured":"Injured","Total Number of Victims":"Total_Victims","Date":"Shooting_Date","Average Shooter Age":"Shooter_Age","Shooter Sex":"Shooter_Gender","Shooter Race":"Shooter_Race"})
mass_1971_1981_rename.head()
```

```python
mass_1971_1981_rename['Shooting_Date'] = pd.to_datetime(mass_1971_1981_rename['Shooting_Date'])
```

```python
mass_1971_1981_rename['Shooting_Date'] = mass_1971_1981_rename['Shooting_Date'].dt.strftime('%Y-%m-%d')
mass_1971_1981_rename.head()
```

```python
start_date = "1971-1-1"
end_date = "1981-1-31"

after_start_date = mass_1971_1981_rename["Shooting_Date"] >= start_date
before_end_date = mass_1971_1981_rename["Shooting_Date"] <= end_date
between_two_dates = after_start_date & before_end_date
filtered_dates = mass_1971_1981_rename.loc[between_two_dates]

filtered_dates.reset_index(inplace=True)
final_1971_1981_df=filtered_dates.drop(columns=['index'])
final_1971_1981_df
```

```python
final_1971_1981_df[['City','State']]=final_1971_1981_df.Shooting_Location.str.split(',',expand=True)
print(final_1971_1981_df)
```

```python
new_df=pd.DataFrame(final_1971_1981_df)
new_df
```

```python
other_df=new_df.drop(columns=['Shooting_Location'])
other_df
```

```python
mass_shooting_1971_1981=other_df[["Shooting_Case","Shooting_Date","City","State","Fatalities","Injured","Total_Victims","Shooter_Age","Shooter_Gender","Shooter_Race"]]
mass_shooting_1971_1981
```
