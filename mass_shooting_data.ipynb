```python
import pandas as pd
import csv
import datetime as dt
from sqlalchemy import create_engine
```

# Read Mass 2016 and Transform Database

```python
csv_file_1="Resources/mass_2016 (1).csv"
mass_1971_1981 = pd.read_csv(csv_file_1)
mass_1971_1981.head()
```

```python
drop_mass=mass_1971_1981[["Title","Location","Number of Victim Fatalities","Number of Victims Injured",
                          "Total Number of Victims","Date","Average Shooter Age","Shooter Sex","Shooter Race"]].copy()
drop_mass.head()
```

```python
mass_1971_1981_rename=drop_mass.rename(columns={"Title":"Shooting_Case","Location":"Shooting_Location",
                                                "Number of Victim Fatalities":"Fatalities",
                                                "Number of Victims Injured":"Injured",
                                                "Total Number of Victims":"Total_Victims",
                                                "Date":"Shooting_Date",
                                                "Average Shooter Age":"Shooter_Age",
                                                "Shooter Sex":"Shooter_Gender",
                                                "Shooter Race":"Shooter_Race"})
mass_1971_1981_rename.head()
```

```python
mass_1971_1981_rename['Shooting_Date'] = pd.to_datetime(mass_1971_1981_rename['Shooting_Date'])
```

```python
mass_1971_1981_rename['Shooting_Date'] = mass_1971_1981_rename['Shooting_Date'].dt.strftime('%Y-%m-%d')
mass_1971_1981_rename.head()
```

```python
start_date = "1971-1-1"
end_date = "1981-1-31"

after_start_date = mass_1971_1981_rename["Shooting_Date"] >= start_date
before_end_date = mass_1971_1981_rename["Shooting_Date"] <= end_date
between_two_dates = after_start_date & before_end_date
filtered_dates = mass_1971_1981_rename.loc[between_two_dates]

filtered_dates.reset_index(inplace=True)
final_1971_1981_df=filtered_dates.drop(columns=['index'])
final_1971_1981_df
```

```python
final_1971_1981_df[['City','State']]=final_1971_1981_df.Shooting_Location.str.split(',',expand=True)
print(final_1971_1981_df)
```

```python
new_df=pd.DataFrame(final_1971_1981_df)
new_df
```

```python
other_df=new_df.drop(columns=['Shooting_Location'])
other_df
```

```python
mass_shooting_1971_1981=other_df[["Shooting_Case","Shooting_Date","City","State","Fatalities","Injured",
                                  "Total_Victims","Shooter_Age","Shooter_Gender","Shooter_Race"]]
mass_shooting_1971_1981
```

# Read Mother Jones DB and Transform Database

```python
csv_file_2="Resources/Mother Jones - Mass Shootings Database, 1982 - 2019 (3).csv"
mass_1982_2021 = pd.read_csv(csv_file_2)
mass_1982_2021.head()
```

```python
drop_mass_2 = mass_1982_2021[['case','location','date','fatalities','injured','total_victims',
                              'age_of_shooter','race','gender']]
drop_mass_2.head()
```

```python
mass_1982_2021_rename=drop_mass_2.rename(columns={"case":"Shooting_Case",
                                                  "location":"Shooting_Location",
                                                  "date":"Shooting_Date",
                                                  "fatalities":"Fatalities",
                                                  "injured":"Injured",
                                                  "total_victims":"Total_Victims",
                                                  "age_of_shooter":"Shooter_Age",
                                                  "race":"Shooter_Race",
                                                  "gender":"Shooter_Gender"})
mass_1982_2021_rename.head()
```

```python
mass_1982_2021_rename['Shooting_Date'] = pd.to_datetime(mass_1982_2021_rename['Shooting_Date'])
```

```python
mass_1982_2021_rename['Shooting_Date'] = mass_1982_2021_rename['Shooting_Date'].dt.strftime('%Y-%m-%d')
mass_1982_2021_rename.head()
```

```python
mass_1982_2021_rename[['City','State']]=mass_1982_2021_rename.Shooting_Location.str.split(',',expand=True)
print(mass_1982_2021_rename)
```

```python
new_df_2=pd.DataFrame(mass_1982_2021_rename)
new_df_2
```

```python
other_df_2 = new_df_2.drop(columns=['Shooting_Location'])
other_df_2
```

```python
mass_shooting_1982_2021=other_df_2[["Shooting_Case","Shooting_Date","City","State","Fatalities",
                                    "Injured","Total_Victims","Shooter_Age","Shooter_Gender","Shooter_Race"]]
mass_shooting_1982_2021
```

# Concat Databases

```python
mass_shootings = pd.concat([mass_shooting_1971_1981, mass_shooting_1982_2021], axis=0)
mass_shootings
```

```python
mass_shootings=mass_shootings.sort_values(['Shooting_Date'])
mass_shootings
```

```python
mass_shootings_db=mass_shootings.reset_index()
final_mass_shootings=mass_shootings_db.drop(columns=['index'])
final_mass_shootings
```

# Connection to SQL

```python
connection_string = "postgres:postgres@localhost:5432/mass_shootings"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
final_mass_shootings.to_sql(name='mass_shootings_1971_2021', con=engine, if_exists='append', index=True)
```
